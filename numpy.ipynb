```python
import numpy as np
from matplotlib import pyplot as plt
x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)
plt.title("Matplotlib demo")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")
plt.plot(x,y,'g1')
plt.show()
```

```python
import numpy as np
from matplotlib import pyplot as plt
x=np.arange(0,3*np.pi,0.1)
y=np.sin(x)
z=np.cos(x)
yy=np.sin(x)
zz=np.cos(x)
plt.title("Matplotlib demo")
plt.xlabel("x axis caption")
plt.ylabel("y axis caption")

plt.subplot(2,4,1)
plt.title("1")
plt.plot(x,y)
plt.subplot(2,4,2)
plt.title("2")
plt.plot(x,z)
plt.subplot(2,4,3)
plt.title("3")
plt.plot(x,yy)
plt.subplot(2,4,4)
plt.title("4")
plt.plot(x,zz)
plt.show()
```
